# Variables dummy

```python
dummy_sex = pd.get_dummies(data['sex'], prefix = 'sex') #Dummificacion
dummy_sex.head()
```

```python
column_name = data.columns.values.tolist()
column_name
```

```python
data.drop(['sex'], axis = 1)
pd.concat([data, dummy_sex], axis = 1) #Agrega las variables dummy al final. 
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df

# Script para dummificar una variable categorica. 
```
